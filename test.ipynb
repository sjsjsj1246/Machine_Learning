```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    ret = 0
    while n > 0:
        print(n)
        ret += (n%10)**2
        n /= 10
        n = int(n)
    return ret
```

```python
x = np.arange(0, 100, 1)
y = np.array([f(n) for n in x])
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y)
plt.xlabel('x', fontsize=15)
plt.ylabel('f(x)', fontsize=15)
plt.axis([0, 100, 0, 170])
plt.grid(True)
plt.xticks(np.linspace(0,100,11))
plt.show()
```
